```python
# import visualizing tools
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.options.display.max_rows = 99
from pandas import DataFrame
import warnings
from pandas.api.types import is_numeric_dtype
```

```python
#read train data
train_df = pd.read_csv('train.csv')
train_df.head()
```

```python
#read test data
test_df = pd.read_csv('test.csv')
test_df.head()
```

```python
# Chekcing the percentage of Null values in all the columns for train
print('Percentage of Missing Values in each column is as follows:')
print(round(train_df.isnull().sum()/len(train_df.index)*100,2).sort_values(ascending=False)[ round(train_df.isnull().sum()/len(train_df.index),2) > 0 ] )
```

```python
# Chekcing the percentage of Null values in all the columns for test
print('Percentage of Missing Values in each column is as follows:')
print(round(test_df.isnull().sum()/len(test_df.index)*100,2).sort_values(ascending=False)[ round(test_df.isnull().sum()/len(test_df.index),2) > 0 ] )
```

```python
# drop columns that are missing more than 10% of training data
new_train_df = train_df.dropna(axis='columns', thresh=train_df.shape[0]*0.9)
(new_train_df.isnull().sum().sort_values(ascending=False)/new_train_df.shape[0])[0:10]
new_train_df = new_train_df.drop(['Id'], axis = 1)
new_train_df.head()
```

```python
# drop columns that are missing more than 10% of test data
new_test_df = test_df.dropna(axis='columns', thresh=test_df.shape[0]*0.9)
(new_test_df.isnull().sum().sort_values(ascending=False)/new_test_df.shape[0])[0:10]
new_test_df = new_test_df.drop(['Id'], axis = 1)
new_test_df.head()
```
